```python
changed_name={'Delhi Daredevils':'Delhi Capitals','Kings XI Punjab':'Punjab Kings','Rising Pune Supergiants':'Rising Pune Supergiant'}
df = df.replace(changed_name.keys(),changed_name.values())
df["Team1"].value_counts()
```

```python
df["Team1"].nunique()
```

```python
df.isna().sum()
df = df.dropna(subset=['WinningTeam'])
df.isna().sum()
```

```python
df
```

```python
teams = df.Team1.unique()
print(teams+"\n")
new_df = pd.DataFrame()
data = []
for team in teams:
    played = df[(df.Team1==team) | (df.Team2==team)].shape[0]
    win = df[df.WinningTeam==team].shape[0]
    win_per = (win/played)*100
    home_win = (df[(df.WinningTeam==team) & (df.Team1==team)].shape[0]/df[df["Team1"]==team].shape[0])*100
    away_win = (df[(df.WinningTeam==team) & (df.Team2==team)].shape[0]/df[df["Team2"]==team].shape[0])*100
    
    data.append([team,played,win_per,home_win,away_win])

new_df[['Team','Match_played','Winning%','Home_Win%','Away_Win']] = data
# print(new_df)
```

```python
new_df
```

```python
teams = df.Team1.unique()
new_df = pd.DataFrame()
data = []
for team in teams:
    played = df[(df.Season=='2022') & ((df.Team1==team) | (df.Team2==team))].shape[0]
    win = df[(df.Season=='2022') & (df.WinningTeam==team)].shape[0]  
    points = win*2
    
    data.append([team,played,win,points])

new_df[['Team_Name','Match_played','Match_Won','Points']] = data
# print(new_df)
```
